In this post, I'll write a tutorial explaining how to construct an interesting data visualization of the *Palmer Penguins* data set.

## Data Acquisition

In this section, you should acquire your data and read it into Python


```python
import pandas as pd
url = "https://raw.githubusercontent.com/PhilChodrow/PIC16B/master/datasets/palmer_penguins.csv"
penguins = pd.read_csv(url)
```

Let's take a look at the data set:

```python
penguins
```

## Make the plot

Next, we can create a histogram using matplotlib:

```python
from matplotlib import pyplot as plt

# Making a histogram grouped by species, of Culmen Depths

fig,ax=plt.subplots(1)
ax.set(xlabel="Culmen Depth (mm)")

# sets species as a set of the three specires of penguins
species= set(penguins['Species'])

for s in species:
    
    # subdivides data by species iterating through set s
    sub=penguins[penguins['Species']==s]
    
    # drops nan values
    sub=sub["Culmen Depth (mm)"].dropna()
    
    ax.hist(sub, label=s.split(' ')[0], alpha=.5)
    ax.legend()
```
